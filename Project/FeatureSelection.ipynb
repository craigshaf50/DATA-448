```python
import boto3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from tqdm import tqdm
from scipy.stats import boxcox
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier 
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.feature_selection import RFE, RFECV

pd.set_option('display.max_columns', 50)

store_final = pd.read_csv('store_final.csv')
store_final.head()
```

# Feature Selection with RFECV #

```python
#defining the input and target variables
x = store_final.drop(columns = ['Response'])
y = store_final['Response']

#defining empty lists to store results
dt_support = []
rf_support = []
ada_support = []
gb_support = []
xgb_support = []
overall_results = []

#repeating 50 times
for i in tqdm(range(0, 50)):
    
    #splitting the data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y)
    
    #RFECV with decision tree
    dt_rfe = RFECV(estimator = DecisionTreeClassifier(max_depth = 3), step = 1, min_features_to_select = 2, 
                   cv = 3, scoring = 'recall', n_jobs = -1).fit(x_train, y_train)
    
    #appending selected features
    dt_support.append(dt_rfe.support_)
    overall_results.append(dt_rfe.support_)
    
    #RFECV random forest
    rf_rfe = RFECV(estimator = RandomForestClassifier(max_depth = 3), step = 1, min_features_to_select = 2, 
                   cv = 3, scoring = 'recall', n_jobs = -1).fit(x_train, y_train)
    
    #appending selected features
    rf_support.append(rf_rfe.support_)
    overall_results.append(rf_rfe.support_)
    
    #RFECV with adaboost
    ada_rfe = RFECV(estimator = AdaBoostClassifier(estimator = DecisionTreeClassifier(max_depth = 3)), step = 1, min_features_to_select = 2, 
                    cv = 3, scoring = 'recall', n_jobs = -1).fit(x_train, y_train)
    
    #appending selected features
    ada_support.append(ada_rfe.support_)
    overall_results.append(ada_rfe.support_)
    
    #RFECV with gradient boosting
    gb_rfe = RFECV(estimator = GradientBoostingClassifier(max_depth = 3), step = 1, min_features_to_select = 2, 
                    cv = 3, scoring = 'recall', n_jobs = -1).fit(x_train, y_train)
    
    #appending selected features
    gb_support.append(gb_rfe.support_)
    overall_results.append(gb_rfe.support_)
    
    #RFECV with xgboost
    xgb_rfe = RFECV(estimator = XGBClassifier(max_depth = 3), step = 1, min_features_to_select = 2, 
                    cv = 3, scoring = 'recall', n_jobs = -1).fit(x_train, y_train)
    
    #appending selected features
    xgb_support.append(xgb_rfe.support_)
    overall_results.append(xgb_rfe.support_)
```

```python
#changing lists to dataframes

#decision tree
dt_support = pd.DataFrame(dt_support, columns = x.columns)
dt = 100 * dt_support.apply(np.sum, axis = 0) /dt_support.shape[0]

#random forest
rf_support = pd.DataFrame(rf_support, columns = x.columns)
rf = 100 * rf_support.apply(np.sum, axis = 0) / rf_support.shape[0]

#adaboost
ada_support = pd.DataFrame(ada_support, columns = x.columns)
ada = 100 * ada_support.apply(np.sum, axis = 0) / ada_support.shape[0]

#gradient boosting
gb_support = pd.DataFrame(gb_support, columns = x.columns)
gb = 100 * gb_support.apply(np.sum, axis = 0) / gb_support.shape[0]

#xgboost
xgb_support = pd.DataFrame(xgb_support, columns = x.columns)
xgb = 100 * xgb_support.apply(np.sum, axis = 0) / xgb_support.shape[0]

#overall results
overall_results = pd.DataFrame(overall_results, columns = x.columns)
results = 100 * overall_results.apply(np.sum, axis = 0) / overall_results.shape[0]
```

```python
#visualizing results
fig, axes = plt.subplots(nrows=3, ncols=2,figsize = (16, 20))

#logistic regression
axes[0, 0].barh(dt.index, dt.values)
axes[0, 0].set_xlabel('Feature Selection Percentage')
axes[0, 0].set_title('Decision Tree')

#random forest
axes[0, 1].barh(rf.index, rf.values)
axes[0, 1].set_xlabel('Feature Selection Percentage')
axes[0, 1].set_title('Random Forest')

#adaboost
axes[1, 0].barh(ada.index, ada.values)
axes[1, 0].set_xlabel('Feature Selection Percentage')
axes[1, 0].set_title('AdaBoost')

#gradient boosting
axes[1, 1].barh(gb.index, gb.values)
axes[1, 1].set_xlabel('Feature Selection Percentage')
axes[1, 1].set_title('Gradient Boosting')

#gradient boosting
axes[2, 0].barh(xgb.index, xgb.values)
axes[2, 0].set_xlabel('Feature Selection Percentage')
axes[2, 0].set_title('XGBoost')

#overall results
axes[2, 1].barh(results.index, results.values)
axes[2, 1].set_xlabel('Feature Selection Percentage')
axes[2, 1].set_title('Overall Results')

fig.tight_layout()
plt.show()
```
