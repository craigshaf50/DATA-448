**Exercise 1: (5 points) Upload the MarketingData.csv data file to your S3 bucket, and using the pandas library, read the MarketingData.csv data file and create a data-frame called marketing_data.**


```python
import boto3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

## Defining the s3 bucket
s3 = boto3.resource('s3')
bucket_name = 'craig-shaffer-data-445-bucket'
bucket = s3.Bucket(bucket_name)

## Defining the file to be read from s3 bucket
file_key = 'MarketingData.csv'
bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

# reading the datafile
marketing_data = pd.read_csv(file_content_stream)
marketing_data.head()
```

**Exercise 3: (15 points) Create two visualizations that may show interesting relationships between the input variables and the target variable. Make sure to describe the visualizations.**

```python
#histograms for each product type
fig, axes = plt.subplots(nrows=2, ncols=3)

ax=axes[0,0]
ax.hist(marketing_data['Fresh'], edgecolor = 'black')
ax.set_title('Fresh')

ax1=axes[0,1]
ax1.hist(marketing_data['Milk'], edgecolor = 'black')
ax1.set_title('Milk')

ax2=axes[0,2]
ax2.hist(marketing_data['Grocery'], edgecolor = 'black')
ax2.set_title('Grocery')

ax3=axes[1,0]
ax3.hist(marketing_data['Frozen'], edgecolor = 'black')
ax3.set_title('Frozen')

ax4=axes[1,1]
ax4.hist(marketing_data['Detergents_Paper'], edgecolor = 'black')
ax4.set_title('Detergents')

ax5=axes[1,2]
ax5.hist(marketing_data['Delicassen'], edgecolor = 'black')
ax5.set_title('Delicassen')

fig.tight_layout()
plt.show()
```

This visualization contains a histogram for each product category. Each product ranges in annual spending from roughly 1000 to 10000 dollars. The distribution for each category is relatively normal. Most histograms peak at around 6000. However, Fresh and Frozen peak higher than the other histograms.

```python
#grouped bar chart based on product category and marketing channel

#finding mean spend for each category based on channel
groups = marketing_data.groupby('Channel').mean()
 
#heights of bars
bars1 = groups.loc[0] #retail
bars2 = groups.loc[1] #road show
bars3 = groups.loc[2] #social media
bars4 = groups.loc[3] #television

barWidth = 0.18

#position of bars on x axis
x1 = np.arange(len(bars1))
x2 = [x + barWidth for x in x1]
x3 = [x + barWidth for x in x2]
x4 = [x + barWidth for x in x3]
 
plt.bar(x1, bars1, width=barWidth, edgecolor='black', label='Retail')
plt.bar(x2, bars2, width=barWidth, edgecolor='black', label='Road Show')
plt.bar(x3, bars3, width=barWidth, edgecolor='black', label='Social Media')
plt.bar(x4, bars4, width=barWidth, edgecolor='black', label='Television')

plt.xlabel('Product')
plt.ylabel('Average Value')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents \nPaper', 'Delicassen'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

Based on the visualization, we can see a that the the Television channel appears to be the best for Delicassen, Frozen, and Milk. It also has the highest average value of all groups with Delicassen with over 7000. The Road Show channel appears to be the best for Fresh and Detergents Paper. However, for the Fresh category, Retail and Social Media aren't far behind. As for the Grocery category, it is led by the Social Media channel. Road Show and Television aren't too far off from leading the Grocery category either. The Retail channel had the lowest average values in four categories (Milk, Grocery, Frozen, and Delicassen).

**Exercise 4: (40 points) Split the data into two data-frames (taking into account the proportion of 0s, 1s, 2s and 3s in Channel): train (80%) and test (20%). Then, do the following:**

```python
#defining inputs
x=marketing_data.drop(columns=['Channel'])
y=marketing_data['Channel']

#splitting the data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .2,stratify=y)
```

(i) Using the one-vs-rest strategy and one model of your preference, build the multi-class classification model. Then, predict the channel label on the test data-frame. Create the classification report.

```python
#building the model
one_vs_rest_rf = OneVsRestClassifier(estimator = RandomForestClassifier(n_estimators = 500, max_depth = 5)).fit(x_train, y_train)

#predicting on test
one_vs_rest_rf_pred = one_vs_rest_rf.predict(x_test)

#printing classification report
print(classification_report(y_test, one_vs_rest_rf_pred))
```

(ii) Using the one-vs-one strategy and the same model from part (i), build the multi-class classification model. Then, predict the channel label on the test data-frame. Create the classification report.

```python
#building the model
one_vs_one_rf = OneVsOneClassifier(estimator = RandomForestClassifier(n_estimators = 500, max_depth = 5)).fit(x_train, y_train)

#predicting on test
one_vs_one_rf_pred = one_vs_one_rf.predict(x_test)

#printing classification report
print(classification_report(y_test, one_vs_one_rf_pred))
```
