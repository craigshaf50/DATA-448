**2. (5 points) How does the Recursive Feature Elimination (RFE) algorithm work? Be specific.**

The RFE algorithm is essentially a backwards selection process of the predictors. First, a model is built with all predictor variables and the corresponding variable importane scores are computed. Then, the least important predictor variable is dropped, another model is built, and the variable importance scores are computed again. The process continues until you are left with optimal variable subset for the specified number of variables to select.

**3. (5 points) Explain k-NN imputation. Be specific.**

In k-NN imputation, missing values are replaced with the mean value from the closest neighbors. The neighbors of each observation are found utilizing Euclidean distance. The replacement value for the missing value can be estimated as the mean or weighted mean of the neighbor’s
value, where neighbors that are further away have less influence on the replacement value.

**6. (5 points) In what scenarios, would you prefer to use the precision-recall curve instead of ROC curve to measure the performance of a classifier? Be specific.**

We would prefer to use the precision-recall curve instead of the ROC curve when the data set is imbalanced. With class imbalanced data, care about decreasing the number of false positive predictions and the precison-recall score does this. In comparison, the ROC curve works best for balanced data sets.

**7. (7 points) Explain the biggest drawback of one-vs-one multi-class classification when compared to one-vs-all multi-class classification. Be specific.**

The biggest drawback for one-vs-one multi-class classification is that it takes more computing time than one-vs-all. This is because the one-vs-all approach only builds a binary classification system for each categorical class while the one-vs-one approach builds a binary classification system for each pair of categorical classes. For example, if you have a dataset with 4 classes one-vs-all would have 4 binary classification systems while one-vs-one would have 6.

**13. (5 points) What is the difference between feature engineering and feature selection? Be specific.**

Feature engineering is the process of transforming existing data into new features that may help uncover relationships in the data and help the model make better predictions. Feature selection is the process of selecting which input variables should be included in the final model. This is important because having a manageable number of inputs reduces computing time and noise from unimportant variables in the model.

**14. (5 points) Based on the discussions from Chapter 5, list at least two benefits of feature selection. Be specific.**

- Having a limited subset of input variables reduces computing time which could be important when building multiple different models
- Including unimportant variables in the model's input variables can hurt the model's overall predictive power. Feature selection helps us decide which variables to keep/use in the model
- By selecting a subset with the most important variables, we can optimize model performance

Profession has 35 missing values. Impute missing values with most frequent label

```python
#imputing missing variables with most frequent label

#identifying categorical variables
categorical_vars = x_train.select_dtypes(include = 'object').columns.to_list()
print(categorical_vars)

#frequent values for each categorical variable (profession is the one we care about)
frequent_values = x_train[categorical_vars].mode().iloc[0].to_dict()
print(frequent_values)

#creating indicator variable for observations with imputed profession (1 for imputed, 0 if not)
x_train['Profession_imputed'] = x_train['Profession'].isna().astype(int)
x_test['Profession_imputed'] = x_test['Profession'].isna().astype(int)

#filling missing values with most frequent labels
x_train = x_train.fillna(value = frequent_values)
x_test = x_test.fillna(value = frequent_values)
```

*(d) (25 points) Using the train data-frame, engineer at least five different features, that can help to predict Spending Score, using the given input variables. Engineer the same features, that you engineer on the train data-frame, on the test data-frame.*

*These are the rules to engineer the features:*
- You can’t use the Box-Cox transformation.
- You can’t use neither 0-1 scaling nor z-score standardization.
- You can engineer at most two features using strong heredity principle. For reference, see homework assignment 4.

```python
#changing Profession to dummies
x_train = pd.concat([x_train.drop(columns=['Profession'],axis=1), pd.get_dummies(x_train[['Profession']])],axis=1)
x_test = pd.concat([x_test.drop(columns=['Profession'],axis=1), pd.get_dummies(x_test[['Profession']])],axis=1)

#changing gender to labels (female-0, male-1)
x_train['Gender'].replace(['Male', 'Female'], [1,0], inplace = True)
x_test['Gender'].replace(['Male', 'Female'], [1,0], inplace = True)
```

Getting feature importances for Strong Heredity Principle features

```python
#defining the empty data frame to store importances
importance = pd.DataFrame(columns = x_train.columns)

for i in range(0, 100):

    #splitting the data
    x_train_i, x_test_i, y_train_i, y_test_i = train_test_split(x_train, y_train, test_size = 0.2, stratify = y_train)
    
    #building the model
    rf_md = RandomForestRegressor(n_estimators = 500, max_depth = 3).fit(x_train_i, y_train_i)

    #extracting the feature importance
    importance.loc[i] = rf_md.feature_importances_.T

#average importance of each of the variables
pd.DataFrame({'Variable':importance.mean().index, 
              'Importance_Score':importance.mean().values}).sort_values(by = 'Importance_Score',
                                                                        ascending = False).reset_index(drop = True)
```

```python
#Features 1 & 2: heredity principle features based on feature importances 
x_train['heredity_feature_1'] = x_train['Annual Income ($)'] * x_train['Age']
x_test['heredity_feature_1'] = x_test['Annual Income ($)'] * x_test['Age']

x_train['heredity_feature_2'] = x_train['Annual Income ($)'] * x_train['Work Experience']
x_test['heredity_feature_2'] = x_test['Annual Income ($)'] * x_test['Work Experience']
```

```python
#Feature 3: Annual Income divided by Family Size
x_train['IncomeOverFamily'] = x_train['Annual Income ($)'] / x_train['Family Size']
x_test['IncomeOverFamily'] = x_test['Annual Income ($)'] / x_test['Family Size']

#Feature 4: Work Experience ^ 2
x_train['squared_WorkExperience']= np.power(x_train['Work Experience'], 2)
x_test['squared_WorkExperience']= np.power(x_test['Work Experience'], 2)

#Feature 5: Annual Income ^ 2
x_train['squared_AnnualIncome']= np.power(x_train['Annual Income ($)'], 2)
x_test['squared_AnnualIncome']= np.power(x_test['Annual Income ($)'], 2)

#Feature 6: natural log of Family Size
x_train['log_FamilySize']= np.log(x_train['Family Size'])
x_test['log_FamilySize']= np.log(x_test['Family Size'])
```

*(e) (20 points) Using the train data-frame (including the engineered features from part (d)), run the RFE algorithm to identify important variables. Run the RFE (using the RFECV function) with a base estimator of your preference, step = 1, min features to select = 2, and cv = 5. Run the RFE 100 times and extract the support of each of the features. Combine the results and rank the features. Tip related to features to be included in the RFE algorithm: if you engineered a feature based only on a single raw input variable, you should only include the engineered feature not both (engineered feature and raw feature).*

```python
rf_support = []
for i in tqdm(range(0, 100)):
    
    #splitting the data
    x_train_i, x_test_i, y_train_i, y_test_i = train_test_split(x_train, y_train, test_size = 0.2, stratify = y_train)
    
    #RFECV with random forest regressor (with 500 trees and max depth of 3)
    rf_rfecv = RFECV(estimator = RandomForestRegressor(max_depth = 3, n_estimators = 500), min_features_to_select=2, cv=3).fit(x_train_i, y_train_i)
    
    #appending selected features
    rf_support.append(rf_rfecv.support_)
```

```python
#random forest features to dataframe
rf_support = pd.DataFrame(rf_support, columns = x_train.columns)
rf = 100 * rf_support.apply(np.sum, axis = 0) / rf_support.shape[0]

plt.barh(rf.index, rf.values)
plt.xlabel('Feature Selection Percentage')
plt.title('Random Forest')
```
